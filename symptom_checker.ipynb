```
smp['Symptom_1'] = smp['Symptom_1'].fillna('')
```

```

# Fill all symptom columns dynamically
symptom_cols = [col for col in smp.columns if col.lower().startswith('symptom')]

# Fill NaNs with empty string
smp[symptom_cols] = smp[symptom_cols].fillna('')

# Cleaning function
def clean_symptom_string(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text)  # normalize spaces
    return text.strip()

# Apply to all symptom columns
for col in symptom_cols:
    smp[col] = smp[col].apply(clean_symptom_string)
```
